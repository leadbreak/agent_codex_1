# Hymba+ 학습 단계 노트북

이 노트북은 학습 단계별 흐름을 간단히 시뮬레이션하고,
손실 변화를 시각적으로 확인하는 용도입니다.

- Stage 1: Pretrain
- Stage 2: SFT
- Stage 3: RL (간단 모사)

```python
import torch
import matplotlib.pyplot as plt

from hymba_plus.core.config import HymbaPlusConfig
from hymba_plus.models.language_model import HymbaPlus

plt.rcParams['font.family'] = ['NanumGothic', 'AppleGothic', 'Malgun Gothic', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
```

```python
config = HymbaPlusConfig.from_yaml('../configs/hymba_plus.yaml')
model = HymbaPlus(config)
model.train()
```

```python
def run_stage(name, steps=5):
    losses = []
    for _ in range(steps):
        input_ids = torch.randint(0, config.vocab_size, (2, 16))
        labels = torch.randint(0, config.vocab_size, (2, 16))
        output = model(input_ids, labels=labels)
        loss = output.loss.item()
        losses.append(loss)
    print(f'{name} 단계 손실:', losses)
    return losses

pretrain_losses = run_stage('Pretrain', steps=5)
sft_losses = run_stage('SFT', steps=5)
rl_losses = run_stage('RL(모사)', steps=5)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(pretrain_losses, label='Pretrain')
plt.plot(sft_losses, label='SFT')
plt.plot(rl_losses, label='RL(모사)')
plt.title('단계별 손실 비교')
plt.xlabel('스텝')
plt.ylabel('손실')
plt.legend()
plt.show()
```
